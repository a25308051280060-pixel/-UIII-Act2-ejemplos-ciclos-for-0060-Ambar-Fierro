```python
# 1. Definición de una Clase y el método
class Estudiante:
    def __init__(self, nombre, edad, nota):
        self.nombre = nombre
        self.edad = edad
        self.nota = nota

    def mostrar_informacion(self):
        print(f"Resultado -> Nombre: {self.nombre}, Edad: {self.edad}, Nota: {self.nota}")

# --- Interacción simple ---
# Pedimos los datos por teclado
nom = input("Escribe el nombre: ")
eda = input("Escribe la edad: ")
not_ = input("Escribe la nota: ")

# Creamos el estudiante con lo que escribió el usuario
alumno = Estudiante(nom, eda, not_)

# Mostramos el resultado
alumno.mostrar_informacion()
```

```python
# 2. Herencia simple
class Persona:
    def __init__(self, nombre, edad):
        self.nombre = nombre
        self.edad = edad

class Empleado(Persona):
    def __init__(self, nombre, edad, sueldo):
        super().__init__(nombre, edad) # Trae lo de Persona
        self.sueldo = sueldo

    def presentarse(self):
        print(f"Hola, soy {self.nombre}, tengo {self.edad} años y gano {self.sueldo}.")

# --- Interacción ---
n = input("Tu nombre: ")
e = input("Tu edad: ")
s = input("Tu sueldo: ")

# Creamos al empleado con tus datos
usuario = Empleado(n, e, s)
usuario.presentarse()
```

```python
# 3. Encapsulamiento con el decorador
class Producto:
    def __init__(self, precio):
        self._precio = precio  # El "_" dice: "No me toques directamente"

    @property
    def precio(self):
        return self._precio

    @precio.setter
    def precio(self, nuevo_precio):
        if nuevo_precio > 0:
            self._precio = nuevo_precio
            print(" Precio actualizado con éxito.")
        else:
            print(" Error: ¡El precio no puede ser 0 o negativo!")

# --- Interacción ---
p = Producto(100)
print(f"Precio inicial: {p.precio}")

nuevo = float(input("Introduce el nuevo precio: "))
p.precio = nuevo  # Esto activa automáticamente el @precio.setter

print(f"Precio final en el sistema: {p.precio}")
```

```python
# 4. Polimorfismo
class Perro:
    def hablar(self):
        return "¡Woof! 🐶"

class Gato:
    def hablar(self):
        return "¡Meow! 🐱"

# --- Interacción ---
print("Elige un animal para escuchar su sonido:")
print("1. Perro")
print("2. Gato")

opcion = input("Escribe 1 o 2: ")

# Polimorfismo en acción:
if opcion == "1":
    animal = Perro()
elif opcion == "2":
    animal = Gato()
else:
    print("Opción no válida")
    animal = None

if animal:
    # No importa si es perro o gato, ambos saben "hablar"
    print(f"El animal dice: {animal.hablar()}")
```

```python
# Abstracción con Clases Abstractas
from abc import ABC, abstractmethod

# La "Plantilla Obligatoria"
class Forma(ABC):
    @abstractmethod
    def calcular_area(self):
        pass

# La clase que cumple el contrato
class Cuadrado(Forma):
    def __init__(self, lado):
        self.lado = lado

    def calcular_area(self):
        return self.lado * self.lado

# --- Interacción ---
print("Cálculo de área de un Cuadrado")
l = float(input("¿Cuánto mide el lado del cuadrado?: "))

# Creamos el objeto
mi_cuadrado = Cuadrado(l)

# Usamos el método obligatorio
print(f"El área resultante es: {mi_cuadrado.calcular_area()}")
```
